```python
import numpy as np
from scipy.stats import kstest
from scipy.stats import levene
from scipy.stats import f_oneway
import scipy.stats as stats
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
from datetime import datetime
```

```python
data=pd.read_excel('/Users/rickie/Desktop/statistics_by_python/sales.xlsx',header=0)
data.head()
```

```python
# data.set_index(keys='Unnamed: 0')
data=data.rename(columns={'Unnamed: 0':'date'})
```

```python
columns=pd.Index(['A','B','C','D'],name='item')
data1=data.reindex(columns=columns)
data1.index=data['date']
data1=data1.stack().reset_index().rename(columns={0:'value'})
data1
```

```python
pd.melt(data,['date'])
```

```python
formula='value~item'
model=ols(formula,data=data1).fit()
model.summary()
```

```python
anova_lm(model)# p<0.05,因此拒绝原假设，得出销量与地区有关的结论
```

```python
tot=data.set_index('date').resample('M',kind='period').sum()
tot=tot[tot.A!=0]
tot=tot.reset_index()

tot.date=tot.date.apply(str)
tot
```

```python
tot1=pd.melt(tot,['date'])
def month_retrieve(x):
    if x=='2019-01':
        return 'Jan'
    if x=='2019-03':
        return 'March'
    if x=='2019-12':
        return 'Dec'
tot1['Month']=tot1.date.apply(month_retrieve)
```

```python
tot1
```

```python
formula='value~Month'
model=ols(formula,tot1).fit()
anova_lm(model)
```

```python
formula='value~Month+variable'
model=ols(formula,tot1).fit()
anova_lm(model)
```

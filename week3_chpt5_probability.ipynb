```python
import numpy as np
from scipy.stats import bernoulli
from scipy.stats import poisson
from scipy.stats import norm
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# 1.伯努利分布
x=np.arange(0,6,1)
p=0.56

plt.plot(x,bernoulli.pmf(x,p),'ro')
```

```python
fig,ax=plt.subplots(1,1)
x = np.arange(bernoulli.ppf(0.01, p),
             bernoulli.ppf(0.99, p))
ax.plot(x, bernoulli.pmf(x, p), 'b.', ms=8, label='bernoulli pmf')
ax.vlines(x, 0, bernoulli.pmf(x, p), colors='b', lw=2, alpha=0.5)
```

```python
#2.泊松分布
mu = 0.6
mean, var, skew, kurt = poisson.stats(mu, moments='mvsk')
```

```python
x=np.arange(poisson.ppf(0.01,mu),poisson.ppf(0.99,mu))
plt.plot(x,poisson.pmf(x,mu),'bo')
```

```python
mu = 1.5
fig, ax = plt.subplots(1, 1)
x = np.arange(poisson.ppf(0.01, mu),poisson.ppf(0.99, mu))
ax.plot(x, poisson.pmf(x, mu), 'co', ms=8, label='poisson pmf')
ax.vlines(x,0,poisson.pmf(x, mu),color='y')
```

```python
plt.bar(x,poisson.pmf(x, mu))
plt.plot(x,poisson.cdf(x,mu))
```

### remark:在n重伯努利实验中，当成功的概率很小（p->0)，实验次数很大时，二项分布接近于伯努利分布

```python
#3.标准正态分布
n=100

x=np.linspace(norm.ppf(0.01),norm.ppf(0.99),100)
plt.plot(x,norm.pdf(x))
```

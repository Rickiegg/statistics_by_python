```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from statsmodels.formula.api import ols

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.preprocessing import StandardScaler
```

```python
data=pd.read_csv('/Users/rickie/Desktop/statistics_by_python/car_sale.csv')
```

```python
data.info()
```

```python
data.head()
```

```python
data.columns=['id','date','sales','x1','x2','x3','x4','x5','x6','x7']
data
```

```python
formula="sales~x1+x2"
model=ols(formula,data).fit()
model.summary()
```

```python
X=data[['x1','x2']]
y=data['sales']
std=StandardScaler()
X=std.fit_transform(X)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25)
lg=LinearRegression()
lg.fit(X_train,y_train)
print('coefficients:',lg.coef_)
print('intercept:',lg.intercept_)
lg.score(X_test,y_test)
```

```python
np.mean(cross_val_score(lg,X_train,y_train,cv=5))
```
